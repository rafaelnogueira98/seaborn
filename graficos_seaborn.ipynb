```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("data\Iris.csv")
iris.head()
```

```python
sns.relplot(x="SepalLengthCm", y="PetalLengthCm", data=iris);
```

```python
sns.relplot(x="SepalLengthCm", y="PetalLengthCm", hue="Species", data=iris);
```

```python
sns.relplot(x="SepalLengthCm", y="PetalLengthCm", hue="Species", style="Species", data=iris);
```

```python
sns.relplot(x="SepalLengthCm", y="PetalLengthCm", size="PetalWidthCm", data=iris);
```

```python
sns.barplot(x="Species", y="SepalLengthCm", data=iris);
```

```python
sns.barplot(y="Species", x="SepalLengthCm", data=iris);
```

```python
iris["Agroupamento_Largura_Petala"] = np.where(iris["PetalLengthCm"] >= 6, "Grande", "Pequena")
iris.head()
```

```python
sns.barplot(x="Species", y="SepalLengthCm", hue="Agroupamento_Largura_Petala", data=iris);
```
